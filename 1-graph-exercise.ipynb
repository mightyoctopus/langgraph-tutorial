```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

## Implement Planning: 
1. Set a state
2. Create a node
3. Build a graph -- add the node and set entry & finish point
4. Invoke an input and run!

```python
class AgentState(TypedDict): 
    """
    Set the state schema for compliment_agent_node
    """
    name: str
```

```python
def compliment_node(state: AgentState) -> AgentState:
    """
    Simple node that updates a name to the state 
    """
    state["name"] = f"Hi {state["name"]}! You're doing an amzing job learning LangGraph!!"

    return state
```

```python
graph = StateGraph(AgentState)
```

```python
graph.add_node("compliment agent", compliment_node)

graph.set_entry_point("compliment agent")
graph.set_finish_point("compliment agent")

app = graph.compile()
```

```python
result = app.invoke({"name": "Bob"})
```
