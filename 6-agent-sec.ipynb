```python
from typing import TypedDict, List
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
CHAT_MODEL = "gpt-5-mini"
```

```python
llm = ChatOpenAI(model=CHAT_MODEL)
```

```python
class AgentState(TypedDict): 
    sys_msg: SystemMessage
    human_msg: HumanMessage
```

```python
def process_llm(state: AgentState) -> AgentState:
    """
    Invoke the llm via API
    """

    res = llm.invoke([state["sys_msg"], state["human_msg"]])
    print(res.content)
    return res
```

```python
graph = StateGraph(AgentState)

graph.add_node("processor", process_llm)

graph.add_edge(START, "processor")
graph.add_edge("processor", END)

app = graph.compile()
```

```python
result = app.invoke({
    "sys_msg": "You are a helpful assistant",
    "human_msg": "what's the best stress reliever of 5 ways. do not exceed 5 lines of text"
})
```
